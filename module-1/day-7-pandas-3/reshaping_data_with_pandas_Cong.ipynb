### Methods for Combining DataFrames: `.join()`, `.merge()`, `.concat()`, `.melt()`

### `.join()`

````python
toy1 = pd.DataFrame([[63, 142], [33, 47], [44, 22]], columns=['age', 'HP'])
toy2 = pd.DataFrame([[63, 100], [33, 200]], columns=['age', 'HP'])
````

````python
toy1
````

````python
toy2
````

````python
toy1.join(toy2.set_index('age'),
          on='age',
          lsuffix='_A',
          rsuffix='_B')
````

````python
toy1.set_index('age').join(toy2.set_index('age'),  #WE are only joining things that match from TOY 2, if TOY 2 has different ages it wont' be joined
                          lsuffix='A')
````

### `.merge()`

````python
ds_chars = pd.read_csv('data/ds_chars.csv', index_col=0)
ds_chars
````

````python
states = pd.read_csv('data/states.csv', index_col=0)
states
````

````python
ds_chars.join(states.set_index('state'),
             on = 'home_state')
````

````python
ds_chars.merge(states,
               left_on='home_state',
               right_on='state',
               how='inner')
# 'inner' vs. left, right, full --> only the join the ones that match, but does NOT overlap
````

### `pd.concat()`

````python
pd.concat([ds_chars, states], sort=False)
````

### `pd.melt()`

Melting removes the structure from your DataFrame and puts the data in a 'variable' and 'value' format.

````python
ds_chars.head()
````

````python
pd.melt(ds_chars,
        id_vars=['name'],
        value_vars=['HP', 'home_state'])
````

````python
pd.melt(animal_outcomes, id_vars=['Animal ID'], value_vars=['Age upon Intake', 'age'])
````
